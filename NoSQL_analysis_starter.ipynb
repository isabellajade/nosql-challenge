```python
# Import dependencies
from pymongo import MongoClient
from pprint import pprint
import pandas as pd
```

```python
# Create an instance of MongoClient
mongo = MongoClient(port=27017)
```

```python
# assign the collection to a variable
establishments = db.establishments
```

### 1. Which establishments have a hygiene score equal to 20?

```python
# Find the establishments with a hygiene score of 20
query = {"scores.Hygiene": 20}

# Use count_documents to display the number of documents in the result
results = establishments.count_documents(query)
print(f'There are {results} documents with a hygiene score of twenty')
# Display the first document in the results using pprint
print("-"*25)
print("The first result is; ")
print("-"*25)
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
hygiene20_df = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
rows = len(hygiene20_df.index)
print(f'There are {rows} rows in this dataframe')
# Display the first 10 rows of the DataFrame
hygiene20_df.head(10)
```

### 2. Which establishments in London have a `RatingValue` greater than or equal to 4?

```python
# Find the establishments with London as the Local Authority and has a RatingValue greater than or equal to 4.
query = {"LocalAuthorityName": {'$regex': "London"},
            "RatingValue": {'$gte':"4"}}

# Use count_documents to display the number of documents in the result
results = establishments.count_documents(query)
print(f'There are {results} documents with London as the Local Authority and has a RatingValue >= 4')

# Display the first document in the results using pprint
print("-"*25)
print("The first result is; ")
print("-"*25)
pprint(establishments.find_one(query))
```

```python
# Convert the result to a Pandas DataFrame
london_df = pd.DataFrame(establishments.find(query))
# Display the number of rows in the DataFrame
rows = len(london_df.index)
print(f'There are {rows} rows in this dataframe')
# Display the first 10 rows of the DataFrame
london_df.head(10)
```

### 3. What are the top 5 establishments with a `RatingValue` rating value of '5', sorted by lowest hygiene score, nearest to the new restaurant added, "Penang Flavours"?

```python
# Search within 0.01 degree on either side of the latitude and longitude.
# Rating value must equal 5
# Sort by hygiene score

degree_search = 0.01
latitude = 51.49014200
longitude = 0.08384000

query = {'RatingValue': '5',
            "geocode.longitude": {'$gte': longitude - degree_search},
                 "geocode.longitude": {'$lte': longitude + degree_search},
                     "geocode.latitude": {'$gte': latitude - degree_search},
                         "geocode.latitude": {'$lte': latitude + degree_search}
        }
sort = [('scores.Hygiene', 1)]
limit = 5

# Print the results
count = establishments.count_documents(query)
print(f'There is a total of {count} documents, here are the first five:')
print("-"*25)

results = list(establishments.find(query).sort(sort).limit(limit))
pprint(results)
```

```python
# Convert result to Pandas DataFrame
results_df = pd.DataFrame(list(establishments.find(query).sort(sort)))

rows = len(results_df.index)
print(f'There are {rows} rows in this dataframe')

results_df.head()
```

### 4. How many establishments in each Local Authority area have a hygiene score of 0?

```python
# Create a pipeline that: 
# 1. Matches establishments with a hygiene score of 0
# 2. Groups the matches by Local Authority
# 3. Sorts the matches from highest to lowest
match_query = {'$match':
                          {'scores.Hygiene': 0}}

group_query = {'$group':
                          {'_id': {'Local Authority': '$LocalAuthorityName'},
                          'count': {'$sum':1}}}

sort_values = {'$sort':
                          {'scores.Hygiene': -1}}

pipeline = [match_query, group_query, sort_values]
pipeline_results = list(establishments.aggregate(pipeline))

# Print the number of documents in the result
print("Number of documents in the results:", len(pipeline_results))
# Print the first 10 results
pprint(pipeline_results[0:10])
```

```python
# Convert the result to a Pandas DataFrame
pipeline_df = pd.json_normalize(pipeline_results)
# Display the number of rows in the DataFrame

# Display the first 10 rows of the DataFrame
pipeline_df.head(10)
```
